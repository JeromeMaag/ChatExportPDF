# Run Local Example Exports

This notebook runs the local synthetic Threema and WhatsApp examples and writes results into `Examples/Output/...`.


```python
from __future__ import annotations

import subprocess
from datetime import datetime
from pathlib import Path

from chat_export import __version__

repo_root = Path.cwd().parent if Path.cwd().name == "Examples" else Path.cwd()
examples_root = repo_root / "Examples"
output_root = examples_root / "Output"
timestamp = datetime.now().astimezone().strftime("%Y-%m-%d_%H-%M-%S")
version = __version__


def run_case(app: str, example_name: str, run_name: str, args: list[str]) -> Path:
    target = output_root / app / example_name / f"{run_name}_{timestamp}_v{version}"
    target.parent.mkdir(parents=True, exist_ok=True)
    command = args + ["--out-dir", str(target)]
    print("RUN:", " ".join(command))
    subprocess.run(command, cwd=repo_root, check=True)
    return target
```

```python
threema_db = examples_root / "ThreemaSynthetic" / "ThreemaData.sqlite"
threema_external = examples_root / "ThreemaSynthetic" / "_EXTERNAL_DATA"
threema_output = run_case(
    "Threema",
    "ThreemaSynthetic",
    "direct_group",
    [
        "python",
        "-m",
        "chat_export",
        "--source",
        "threema",
        "--db-path",
        str(threema_db),
        "--external-folder",
        str(threema_external),
    ],
)
threema_output
```

```python
wa_direct_zip = (
    examples_root / "WhatsAppSynthetic" / "WhatsApp-Chat mit Eva Freitag.zip"
)
wa_direct_output = run_case(
    "WhatsApp",
    "WhatsAppSynthetic_Direct_iOS",
    "ios_direct",
    [
        "python",
        "-m",
        "chat_export",
        "--source",
        "whatsapp",
        "--input-path",
        str(wa_direct_zip),
    ],
)
wa_direct_output
```

```python
wa_group_zip = (
    examples_root / "WhatsAppSynthetic" / "WhatsApp Chat - Projektteam Orion.zip"
)
wa_group_output = run_case(
    "WhatsApp",
    "WhatsAppSynthetic_Group_Android",
    "android_group",
    [
        "python",
        "-m",
        "chat_export",
        "--source",
        "whatsapp",
        "--input-path",
        str(wa_group_zip),
    ],
)
wa_group_output
```

```python
print("Created outputs:")
print("-", threema_output)
print("-", wa_direct_output)
print("-", wa_group_output)
```
